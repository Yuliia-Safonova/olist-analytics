# Part 3. Python: Simple Forecast and Recommendations

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

RANDOM_STATE = 42

FILE_IN = 'https://raw.githubusercontent.com/Yuliia-Safonova/olist-analytics/refs/heads/main/data/main_sales_reg.csv'
```

```python
df = pd.read_csv(FILE_IN)
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df = df.sort_values('ym').reset_index(drop=True)
print('Month (all):', len(df))
df.head()
```

```python
df = df[(df.ym >= '2017-01') & (df.ym <= '2018-08')].reset_index(drop=True)
print('Month (in process):', len(df))
display(df.head(2))
df.tail(2)
```

```python
df.isna().sum()
```

```python
pivot = df.pivot_table(
    values='price',
    index='ym',
    aggfunc='sum'
)

pivot
```

```python
plt.figure(figsize=(9, 4))
plt.plot(pivot.index, pivot.price, marker="o")
plt.xticks(rotation=45, ha='right')
plt.title('Monthly revenue (stable period)')

plt.tight_layout()
plt.show()
```

```python
df_new = {
    't': list(range(len(pivot))),
    'ym': list(pivot.index),
    'revenue': list(pivot.price)
}
df_new = pd.DataFrame(df_new)
df_new
```

```python
X = df_new[['t']]
y = df_new['revenue']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=RANDOM_STATE)

model = LinearRegression().fit(X_train, y_train)

next_index = len(df_new)
y_pred = model.predict(X_test)

forecast = model.predict(pd.DataFrame({
    't': [next_index]
}))[0]
print(f"Predict revenue for next month: R${forecast:,.0f}")
print(f"Average growth for next month: R${model.coef_[0]:,.0f}")
print('-' * 15)

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print(f'MAE: R${mae:,.0f}')
print(f'RMSE: R${rmse:,.0f}')
print(f'Predict accuracy (R2): {r2:.1%}')
print('-' * 15)

print(f"Predict revenue for next month: R${forecast:,.0f} ± {mae:,.0f}")
```

```python
df_new['month'] = df_new['ym'].str[-2:]
season = df_new.groupby('month')['revenue'].mean().round(0)
print('Average growth for number month:')
print(season.sort_values(ascending=False))
```
